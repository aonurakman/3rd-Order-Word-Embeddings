```python
import matplotlib.pyplot as plt
```

```python
colors = ['#2B2F6C', '#564779', '#4C6A78', '#C17F8B', '#532200', '#22404A']
model_names = ["base (17M)", "v15 (17M)", "v18 (17M)", "v26 (17M)", "word2vec (100B)", "GloVe (6B)"]
short_names = ["base", "v15", "v18", "v26", "w2v", "glove"]
```

```python
model_idx = 5
disp_color = colors[model_idx]
model_name = model_names[model_idx]
short_model_name = short_names[model_idx]
```

```python
# Read the data file
with open('wordsim_similarity.txt', 'r') as file:
    lines = file.readlines()

# Initialize empty lists to store data
model_predictions = []

# Process each line and extract model predictions
for line in lines:
    # Extract the model prediction
    model_prediction = float(line.split('[Model: ')[1].split(']')[0])
    model_predictions.append(model_prediction)

# Set the histogram bins and create the histogram
bins = [i for i in range(11)]  # Divide the range into 0.1 intervals
plt.hist(model_predictions, bins=bins, color=disp_color, edgecolor='k', alpha=0.7)

# Customize the plot
plt.xlabel('Model Predictions')
plt.ylabel('Frequency')
plt.title('Histogram of %s Predictions (%s)' % ("WS (Sim)", model_name))

plt.savefig('%s_predictions_(%s)' % ("ws-sim", short_model_name))

# Display the histogram
plt.show()
```
